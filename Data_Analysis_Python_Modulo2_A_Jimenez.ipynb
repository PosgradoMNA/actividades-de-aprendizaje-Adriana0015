4. Data Normalization in Python

Importante tecnica para entender preprocesamiento de datos.

Esto facilita el tener datos uniformes.

Al momento de tener datos muy diferentes el dato de mayor nomenclatura puede crear bias al momento de aplicar modelos de regresion, ya que este pesara mas que los demas datos. Eso no significa que sea mas importante para el analisis del modelo.
Es por eso que es importante normalizar los datos!

3 tecnicas
1) Simple feature scaling> xnew=xold/xmax
2) Min-max> xnew=xold-xmin/xmax-xmin
3) Z-score > xnew=xold-miu/std dev
