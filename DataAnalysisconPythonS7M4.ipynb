**4.2 Linear Regression and Multiple Linear Regression**

Regresion Lineal

-Se refiere a una variable independiente para hacer una prediccion.

Regresion Lineal Multiple

-Se refiere a multiples variables independientes para hacer una prediccion.

Regresion Lineal Simple> es un metodo que se utiliza para entender la relacion entre dos variables: el predictor (independiente) o variable x y el objetivo (dependiente) o variable y
*y= b0 + b1 x*

Para poder entender el modelo lineal, se pueden tomar varios puntos de la recta. Cuando se usa los punto de entrenamiento para ajustar el modelo, se dice que los resultados de los puntos de entrenamiento son los parametros.

Se pueden almacenar los puntos de datos en dos dataframes o arreglos numpy.

El valor que se desea predecir se llama el objetivo, que se almacena en el arreglo como y. Almacenamos la variable dependiente en el dataframe como arreglo X



---



En conclusion:

Se tienen sets de puntos de entrenamiento> estos se usan para entrenar el modelo y obtener parametros.

Esos parametros son usados en el modelo, ahora que se tiene el modelo se utiliza en la variable y que nos da un modelo estimado. 

Este modelo puede ser usado para predicir los valores que no se han visto.

```python
df= pd.read_csv('/content/sample_data/default of credit card clients.csv') 
df

cdf=df.copy() #para no modificar original
cdf
```

```python
#como ajustar un modelo lineal simple
from sklearn.linear_model import LinearRegression

#crear objeto de regresion lineal
lm=LinearRegression()
```

```python
#definimos la variable predictor y la variable de objetivo
X= cdf [['X1']]
Y= cdf[['X15']]

#ahora se usa el objeto lm para ajustar el modelo, esto es igual que usar los parametros b0 y b1
lm.fit(X,Y)

#de aqui se puede obtener la prediccion, usando el metodo predict
Yhat=lm.predict(X)
```

```python
#MLR Regresion Lineal Multiple
#este es en un plano de 2D

#se puede extraer 4 variables predictores y almacenarlos en una variable

z= cdf [['X1','X3','X4','X5']]

#ahora se entrena el modelo

lm.fit(z, cdf['X20'])

#hacer prediccion, con metodo predict
Yhat=lm.predict(X)
```
