# 6. Escala los datos, si consideras necesario


```python
#escalar los datos
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled = scaler.fit_transform(df_cat)

scaled[:5]
```

```python
scaled_df_cat = pd.DataFrame(scaled, columns=df_cat.columns)
scaled_df_cat.head()
```

# 7. Reduce las dimensiones con PCA, si consideras necesario.
7a.Indica la varianza de los datos explicada por cada componente seleccionado.Para actividades de exploración de los datos la varianza > 70%
7b.Indica la importancia de las variables en cada componente


```python
from sklearn.decomposition import PCA
pcs = PCA()
pcs_t = pcs.fit_transform(scaled_df_cat)
pcs_t[:5]
```

```python
pcsSummary_ncdf2 = pd.DataFrame({
'% varianza explicada': np.round(pcs.explained_variance_ratio_,4) * 100,
'% varianza acumulada': np.cumsum(pcs.explained_variance_ratio_) * 100
})
pcsSummary_ncdf2
```

```python
pcs_labels = [f'PC{i + 1}' for i in range(len(scaled_df_cat.columns))]
pcsSummary_ncdf2.index = pcs_labels
pcsSummary_ncdf2
```
